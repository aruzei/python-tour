```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
x = np.arange(-3.2,3.2,0.1)
sinx  = np.sin(x)
cosx = np.cos(x)
```

```python
plt.plot(x,sinx,label='sin x')
plt.plot(x,cosx,linestyle='--',label='cos x')
plt.xlabel("x")
plt.ylabel("y")

plt.legend()

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          ['$-\pi$', '$-\pi/2$', '$0$', '$+\pi/2$', '$+\pi$'])

plt.yticks([-1, 0, +1],
          ['$-1$', '$0$', '$+1$'])

plt.show()
```

```python
imgA = imread('images/3.png')
imgB = imread('images/6.png')
imgC = imread('images/9.png')
```

```python
def showimages(images):
    for zero_based_index,img  in enumerate(images):
        plt.subplot(1,len(images),zero_based_index+1)
        plt.imshow(img)
        
    plt.show()

showimages([imgA,imgB,imgC])

```
